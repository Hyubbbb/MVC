```python
# Define Weekday function

def find_friday(article_date):
    # Monday = 0 & Sunday = 6
    current_weekday = article_date.weekday()

    # Days until next Friday (4) from current date
    days_until_this_friday = (4 - current_weekday) % 7
    this_friday = article_date + timedelta(days=days_until_this_friday)

    return this_friday


# TEST
article_date = datetime(2024, 8, 2)

fridays = find_friday(article_date)

print("이번주: " + find_friday(article_date).strftime('%Y-%m'))
print("다음주: " + (find_friday(article_date) + timedelta(days=7)).strftime('%Y-%m'))
print("지난주: " + (find_friday(article_date) - timedelta(days=7)).strftime('%Y-%m'))
```

```python
def reformat_dates(text):
    # Regular expression to match dates in YYYY-MM-DD or YYYY-MM format
    pattern = r'\d{4}-\d{2}(?:-\d{2})?'

    # Function to format matched date strings
    def format_match(match):
        date_str = match.group(0)
        try:
            # Check if it includes a day
            if '-' in date_str[7:]:  # Looks for the second '-' in 'YYYY-MM-DD'
                date_obj = datetime.strptime(date_str, '%Y-%m-%d')
                return date_obj.strftime('%Y년 %m월 %d일')
            else:
                date_obj = datetime.strptime(date_str, '%Y-%m')
                return date_obj.strftime('%Y년 %m월')
        except ValueError as e:
            # In case of a parsing error, return the original date string
            return date_str

    # Replace all found dates with their new format
    return re.sub(pattern, format_match, text)
```
